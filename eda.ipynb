# Eurovision Song Contest Data

The dataset contains voting and polling data for the six most recent editions of the Eurovision Song Contest (2022, 2021, 2019, 2018, 2017, and 2016).

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
```

## Voting Data

The dataset contains data on how different countries allocated their points in Eurovision finals.

Each participating country awards two sets of **12**, **10**, **8â€“1** points to their ten favourite songs: one set from their professional jury and the other from televoting.

### Jury Data

Each cell contains the number of points awarded by the column country's jury to the row country.

```python
def get_jury_data(year):
    path = f"./data/Final Results/Jury/{year}_jury_results.csv"
    df = pd.read_csv(path, index_col="Contestant")
    df = df.iloc[:, 3:]
    df = df.fillna(0)
    df = df.astype("int32")
    return df


jury_data = get_jury_data(2022)

jury_data
```

### Televote Data

Each cell contains the number of points awarded by the column country's public to the row country.

```python
def get_televote_data(year):
    path = f"./data/Final Results/Televote/{year}_televote_results.csv"
    df = pd.read_csv(path, index_col="Contestant")
    df = df.iloc[:, 3:]
    df = df.fillna(0)
    df = df.astype("int32")
    return df


televote_data = get_televote_data(2022)

televote_data
```

## Polling Data

### Eurovisionworld Poll

Every year [Eurovisionworld](https://eurovisionworld.com/) carries out a poll where users can select which country they believe will win Eurovision.

Users can only vote once for a single country. The dataset contains the number of votes awarded to each country.

```python
def get_eurovision_world_data(year):
    path = f"./data/Polls/Eurovision World Poll/{year}_eurovisionworld_results.csv"
    df = pd.read_csv(path, index_col="Contestant")
    return df


eurovision_world_data = get_eurovision_world_data(2022)

eurovision_world_data
```

### Organisation GÃ©nÃ©rale des Amateurs de l'Eurovision (OGAE) Poll

Every year [OGAE](https://ogaeinternational.org/) carries out a poll where users can award points to the countries of their top 10 songs (akin to the Eurovision points system).

The dataset contains the average number of points awarded to each country.

```python
def get_ogae_data(year):
    path = f"./data/Polls/OGAE Poll/{year}_ogae_results.csv"
    df = pd.read_csv(path, index_col="Contestant")
    return df


ogae_data = get_ogae_data(2022)

ogae_data
```

## Contest Data

This data contains basic information about each edition of the contest.

```python
def get_contest_data():
    path = "./data/contest_data.csv"
    df = pd.read_csv(path, index_col="year")
    df = df.fillna(0)
    df = df.astype(
        {
            "jury_countries_voting_final": "int32",
            "televote_countries_voting_final": "int32",
        }
    )
    return df


contest_data = get_contest_data()

contest_data
```

## Country Data

This data contains the geographic region of each country.

```python
def get_country_data():
    path = "./data/country_data.csv"
    df = pd.read_csv(path, index_col="country")
    return df


country_data = get_country_data()

country_data
```

## Song Data

This data contains information about each entrant.

```python
def get_song_data():
    path = "./data/song_data.csv"
    df = pd.read_csv(path, encoding="ISO-8859-1", index_col=["year", "country"])
    return df


song_data = get_song_data()

song_data
```

We plot some distributions of the columns.

We also see that we have some missing data in some of the columns, and so this will need to be taken into account when deciding what we can visualise. We cannot fill in the missing data in traditional ways (such as with median values), so we leave it as is. Ultimately we may decide to not use data where we are missing too many values.

Besides this one file, the data is perfectly clean and we may not need to use this file (depending on the scope of our visualisation).
